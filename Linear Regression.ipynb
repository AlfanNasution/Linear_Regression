```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('cost_revenue_clean.csv')
df
```

```python
df.describe()
```

```python
2.78e+09
```

```python
x = pd.DataFrame(df, columns=['production_budget_usd'])
y = pd.DataFrame(df, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.title('Prodution Budget vs Worldwide Gross')
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
```

```python
regression = LinearRegression()
```

```python
regression.fit(x,y)
```

Slope Coefficient:

```python
regression.coef_ #theta_1
```

```python
#intercept
regression.intercept_
```

```python
#Plot the Linear Regression into The Canvas

plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.plot(x, regression.predict(x), color='r', lw=3)

plt.title('Prodution Budget vs Worldwide Gross')
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
```

```python
regression.score(x,y)
```
